```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
df =remove_outliers(df,'JamsLength')
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
df =remove_outliers(df,'JamsDelay')
```

```python
severity_counts = df.groupby(['City', 'TrafficIndexLive']).size().unstack()

# Plotting the bar chart
severity_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of City by TrafficIndexLive')
plt.xlabel('City')
plt.ylabel('TrafficIndexLive')
plt.xticks(rotation=45)
plt.legend(title='TrafficIndexLive', bbox_to_anchor=(1.0, 1.0))
plt.show()
```
